```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
sns.set_theme()
```

# Plotting

```python
df = pd.read_csv('mpg.csv')
df['model'] = df.name.str.split(' ').str[0]
```

```python
sns.relplot(data=df, x='weight', y='mpg', hue='origin'); # Scatter plot
# sns.scatterplot(data=df, x='weight', y='mpg', hue='origin'); # Scatter plot
```

```python
sns.lmplot(data=df, x='weight', y='mpg', hue='origin');
```

```python
sns.relplot(data=df, x='model_year', y='mpg', hue='origin', kind='line'); # Line plot
# sns.lineplot(data=df, x='model_year', y='mpg', hue='origin'); # Line plot
```

```python
sns.displot(data=df, x='mpg', hue='origin', kde=True); # Histogram plot with KDE
# sns.histplot(data=df, x='mpg', hue='origin', kde=True); # Histogram plot with KDE
```

```python
sns.displot(data=df, x='mpg', hue='origin', kind='kde'); # KDE Plot
# sns.kdeplot(data=df, x='mpg', hue='origin'); # KDE Plot
```

```python
sns.displot(data=df, x='mpg', y='model_year', hue='origin', kind='hist', col='origin'); # KDE Plot
```

```python
sns.catplot(data=df, x='origin', y='mpg', hue='cylinders', kind='bar'); # Bar plot
# sns.barplot(data=df, x='origin', y='mpg', hue='cylinders'); # Bar plot
```

```python
sns.catplot(data=df, x='origin', y='mpg', hue='cylinders', kind='box'); # Box plot
# sns.boxplot(data=df, x='origin', y='mpg', hue='cylinders'); # Box plot
```

```python
sns.jointplot(data=df, x='weight', y='mpg', hue='origin'); # Joint distribution plot
```

```python
sns.pairplot(data=df[['mpg','weight','acceleration','origin']], hue='origin'); # Pairwise joint distribution plot
```

```python
sns.heatmap(df.corr(), cmap='PRGn');
```
